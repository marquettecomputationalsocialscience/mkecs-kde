```python
lessThanHSDiploma = survey_data[survey_data['Q34'] == "Less than high school diploma"]
ged = survey_data[survey_data['Q34'] == "High school diploma or GED"]
someCollege = survey_data[survey_data['Q34'] == "Some college"]
currentlyEnrolledCollege = survey_data[survey_data['Q34'] == "Currently enrolled in college"]
bachelors = survey_data[survey_data['Q34'] == "Bachelor's degree"]
masters = survey_data[survey_data['Q34'] == "Master's degree"]
doctorate = survey_data[survey_data['Q34'] == "Doctorate degree"]

educationCounts = pd.DataFrame({'counts': [len(lessThanHSDiploma), len(ged), len(someCollege),
                                          len(currentlyEnrolledCollege), len(bachelors), len(masters),
                                          len(doctorate)]},
                              index=['less than HS diploma', 'GED', 'some college', 'currently enrolled in college',
                                    'bachelors', 'masters', 'doctorate'])
educationCounts.plot.bar(y='counts', figsize=(6,6))
```

```python
white = survey_data[survey_data['Q32'] == "White"]
africanAmerican = survey_data[survey_data['Q32'] == "Black or African American"]
eastAsian = survey_data[survey_data['Q32'] == "East Asian"]
southAsian = survey_data[survey_data['Q32'] == "South Asian"]
pacificIslanderHawaiian = survey_data[survey_data['Q32'] == "Pacific Islander or Hawaiian"]
middleEastern = survey_data[survey_data['Q32'] == "Middle Eastern or Arab"]
nativeAmerican = survey_data[survey_data['Q32'] == "Native American or First Nations"]
hispanic = survey_data[survey_data['Q32'] == "Central American Native, South American Native, Chicano, or Mestizo"]
other = survey_data[survey_data['Q32'] == "Other (please specify)"]

raceCounts = pd.DataFrame({'counts': [len(white), len(africanAmerican), len(eastAsian),
                                     len(southAsian), len(pacificIslanderHawaiian), len(middleEastern),
                                     len(nativeAmerican), len(hispanic), len(other)]},
                         index=['white', 'african american', 'east asian', 'south asian', 'pacific islander or hawaiian',
                               'middle eastern', 'native american or first nations', 'central american native, south american native, chicano, or mestizo',
                               'other'])
raceCounts.plot.bar(y='counts', figsize=(6,6))
```
